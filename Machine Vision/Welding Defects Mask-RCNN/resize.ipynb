```python
# 調整圖片大小
import glob
from PIL import Image

imgs = glob.glob('./dataset/total/*.png')
for i in imgs:
    print(i)
    im = Image.open(i)
    size = im.size
    im = im.convert("RGB")
    name = i.split('\\')[::-1][0]        # 取得圖片的名稱
    im2 = im.resize((1024, 750))         # 調整圖片尺寸為 200x200
    im2.save(f'./dataset/resize/total/{name}')
```
