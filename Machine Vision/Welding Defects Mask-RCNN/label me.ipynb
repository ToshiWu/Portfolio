```python
import os
import shutil
```

```python
# 整理編號
def reset_name(workpath):
    for n, i in enumerate(os.listdir(workpath)):
        filename = ('0' if n//2+1<10 else '') + str(n//2+1)+'.'+ i.split('.')[1]
        print(i+'\t-> '+filename)
        os.rename(workpath+i,workpath+filename)
```

```python
# 創建實例
def labelme_json_to_dataset(workpath):
    for i in os.listdir(workpath):
        if i[-5:] == '.json':
            if '_'.join(i.split('.')) not in os.listdir(workpath):
                print(i)   
                os.system("labelme_json_to_dataset " + workpath + i)
```

```python
# 整理資料
def copy_mask(workpath, mask_path):
    cwd = os.path.abspath(os.getcwd())
    for i in os.listdir(workpath):
        filepath = os.path.join(workpath, i)
        if os.path.isfile(os.path.join(filepath, 'label.png')):
            source_file = os.path.join(filepath, 'label.png')
            destination_file = os.path.join(mask_path, i.split('_')[0]+'.png')
            shutil.copyfile(source_file, destination_file)
    os.chdir(cwd)        
```

```python
workpath = './dataset/resize/train/'
mask_path = './train_data/cv2_mask/'
reset_name(workpath)
labelme_json_to_dataset(workpath)
copy_mask(workpath, mask_path)

workpath = './dataset/resize/val/'
mask_path = './val_data/cv2_mask/'
```
