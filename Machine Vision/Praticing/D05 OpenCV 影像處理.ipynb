
使用 OpenCV 進行影像標注,使用 matplotlib 繪製多圖。

1. 紅匡標注、對角線繪製、座標標注、多圖繪製
2. 取得 圖片的 histogram

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./D00_lena.png')
```

### 第一大題 (OpenCV 標註訓練)

1. 將 Lena 的臉蛋用紅框框起來
2. 畫出紅匡中的對角線
3. 標注紅匡左上角的座標
4. 利用 subplot 將三張圖片繪製出來

```python
# 獲得用紅框標注的圖片
img_boxed = img.copy()
cv2.rectangle(img_boxed, (60, 40), (420, 510), (0, 0, 255), 3)

# 獲得用對角線標注的圖片
img_diagonal = img_boxed.copy()
cv2.line(img_diagonal, (60, 40), (420, 510), (0, 0, 255), 3)

# 獲得具有座標標注的圖片
img_coord = img_boxed.copy()
cv2.putText(img_coord, '(60, 40)', (60, 30), 0, .8, (0, 0, 0), 2)

# 利用 subplot 將三張圖片繪製出來
plt.figure(figsize = (15,5))
for i,x in enumerate([img_boxed, img_diagonal, img_coord]):
    plt.subplot(1, 3, i+1).imshow(x[..., [2, 1, 0]])
    plt.axis('off')
plt.show()
```

### 第二大題 (繪製 Histogram)

1. 獲得 3 個 channel (RGB) 的 histogram
2. 利用 subplot 繪製出來

```python
# 繪製 Histogram
plt.figure(figsize = (12,15))
for i in range(3):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.subplot(3, 1, i+1, title = f'Channel {i}').bar(range(0, 256), hist[:, 0])
plt.show()
```
