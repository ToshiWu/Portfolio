# Gaussian mixture model (GMM) or EM expectation-maximization clustering 

```r
library(mclust) # for gmm
```

```r
data("iris")
result.gm <- Mclust(iris[,-5], G=3)

# 輸出詳細資訊
print(result.gm$parameter)
# 輸出分類結果
result.gm$classification
```

```r
# Gaussian mixture model聚類的效果
table(iris[,5],result.gm$classification)
```

```r
# 可視化(因為指定群數,無BIC趨勢)
plot(result.gm) # (1:BIC, 2:classification, 3:uncertainty, 4:density, 0:exit)
```

```r
# 分群結果可視化
mclust2Dplot(iris[, 3:4], classification=result.gm$classification)
```

```r
# 預設會分析分群數
fitGM1 <- Mclust(iris[,-5])
summary(fitGM1, parameters=TRUE)
plot(fitGM1)
```

```r
# BIC 結果
fitGM1$BIC # 輸出相等於使用 mclustBIC()
```

```r
# 分群結果可視化
mclust2Dplot(iris[, 3:4], classification=fitGM1$classification)
```

```r
# 2D密度可視化 (以petal為例)
iris.petal <- densityMclust(iris[, 3:4])
plot(iris.petal, iris[, 3:4], nlevels=15)
```

```r
# 3D密度可視化 (以sepal為例)
iris.sepal <- densityMclust(iris[, 1:2])
plot(iris.sepal, iris[, 1:2], what = "density", type="persp", col = 2)
```

# 非正規化性別資料集

```r
# 讀入資料GMM聚類分析
gender_size <- read.csv("../data_files/gender_size.csv")
fitGM2 <- Mclust(gender_size[,-4])
summary(fitGM2, parameters=TRUE)
```

```r
# 最佳結果分為三群
print(fitGM2$parameters)
fitGM2 $ classification
```

```r
# Gaussian mixture model聚類的效果
classification <- fitGM2$classification
table(gender_size$Gender, classification)
```

```r
# 可視化結果
plot(gender_size[,1:2], pch=classification, col=classification)
plot(gender_size[,2:3], pch=classification, col=classification)
plot(gender_size[,1], gender_size[,3],
     xlab=names(gender_size[1]), ylab=names(gender_size[3]),
     pch=classification, col=classification)

# 以上結果為下圖的下三角形部分
plot(gender_size[,1:3], pch=classification, col=classification)
```
