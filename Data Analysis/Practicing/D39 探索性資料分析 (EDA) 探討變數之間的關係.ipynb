```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import pingouin as pg
import researchpy
from scipy import stats
```

## è®€å…¥è³‡æ–™

```python
df_train = pd.read_csv("./D36-41 æ¢ç´¢æ€§è³‡æ–™åˆ†æ (EDA)_Dataset/Titanic_train.csv")
print(df_train.info())
```

```python
# é€™é‚Šæˆ‘å€‘åšä¸€å€‹èª¿æ•´ï¼ŒæŠŠ Survived è®Šæˆé›¢æ•£å‹è®Šæ•¸ Survived_cate
df_train['Survived_cate'] = df_train['Survived'].astype('object')
print(df_train.info())
```

###  Q1: é€éæ•¸å€¼æ³•è¨ˆç®—  Age å’Œ  Survived  æ˜¯å¦æœ‰ç›¸é—œæ€§?


```python
# é€£çºŒå‹ vs é›¢æ•£å‹
# å–å‡ºç„¡éºå¤±å€¼çš„æ•¸æ“š
data = df_train[['Age','Survived_cate']].dropna()
display(data)
```

```python
# é‹ç”¨ pg.anova è¨ˆç®—ä¸‰ç¨®è®Šç•°æ•¸
aov = pg.anova(dv='Age', between='Survived_cate', data=data, detailed=True)
display(aov)
```

```python
# è®Šç•°æ•¸æ›ç®—å¾—åˆ° Eta Squared (ğœ¼^ğŸ)
etaSq = aov.SS[0] / (aov.SS[0] + aov.SS[1])
print('etasq:', etaSq)
```

```python
def judgment_etaSq(etaSq):
    for condition in {'Negligible': .01,
                      'Small': .06,
                      'Medium': .14}.items():
        if etaSq < condition[1]:
            return condition[0]
    return 'Large'
judgment_etaSq(etaSq)
```

```python
plt.rcParams["figure.dpi"] = 150
g = sns.catplot(x="Survived_cate", y="Age", hue="Survived_cate", data=data, kind="violin")
plt.show()
```

### Q2:é€éæ•¸å€¼æ³•è¨ˆç®—  Sex å’Œ Survived  æ˜¯å¦æœ‰ç›¸é—œæ€§?


```python
# é›¢æ•£å‹ vs é›¢æ•£å‹
# ç”¨äº¤å‰åˆ—é€£è¡¨(contingency table)ï¼Œä¾†æ•´ç†å…©å€‹é¡åˆ¥å‹çš„è³‡æ–™
contTable = pd.crosstab(df_train['Sex'], df_train['Survived_cate'])
display(contTable)

# è¨ˆç®—è³‡æ–™è‡ªç”±åº¦ df*
df = min(contTable.shape[0], contTable.shape[1]) - 1
print(df)
```

```python
# é‹ç”¨ researchpy å¥—ä»¶ï¼Œè¨ˆç®—å‡º Cramerâ€™s V ä¿‚æ•¸
# é€™é‚Šç”¨å¡æ–¹æª¢å®šç¨ç«‹æ€§ï¼Œæ‰€ä»¥æ¡ç”¨çš„ test åƒæ•¸ç‚ºå¡æ–¹ "test =" argument.
# æ¡ç”¨çš„è®Šæ•¸åœ¨é€™å€‹æ¨¡çµ„ä¸­ï¼Œæœƒè‡ªå·±æ ¹æ“šè³‡æ–™é›†ä¾†åˆ¤æ–·ï¼ŒCramer's Phi if it a 2x2 table, or Cramer's V is larger than 2x2.
rosstab, res = researchpy.crosstab(df_train['Survived_cate'], df_train['Sex'], test='chi-square')
CramerV = res.loc[2, 'results']
print(res, "\n\nCramer's value is", CramerV)
```

```python
## å¯«ä¸€å€‹å‰¯ç¨‹å¼åˆ¤æ–·ç›¸é—œæ€§çš„å¼·åº¦
def judgment_CramerV(df, V):
    df = df if df in range(1,5) else 5
    for case in {1: {'Negligible': .1,
                     'Small': .3,
                     'Medium': .5},
                 2: {'Negligible': .07,
                     'Small': .21,
                     'Medium': .35},
                 3: {'Negligible': .06,
                     'Small': .17,
                     'Medium': .29},
                 4: {'Negligible': .05,
                     'Small': .15,
                     'Medium': .25},
                 5: {'Negligible': .05,
                     'Small': .13,
                     'Medium': .22}}.items():
            for condition in case[df].items():
                if V < condition[1]:
                    return condition[0]
    return 'Large'

print(judgment_CramerV(df, CramerV))
```
