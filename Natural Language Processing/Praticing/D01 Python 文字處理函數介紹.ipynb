## 讀入所需文件

```python
import pandas as pd
#TSV文件有別於CSV的文件是使用\ t作為分隔符，CSV則是使用，作為分隔符。
dataset = pd.read_csv('./D00_Restaurant_Reviews.tsv', sep='\t')
all_review = dataset['Review'].values
```

---

### 計算有多少個句子是以 . 結尾

```python
n=[1 if i.endswith('.') else 0 for i in all_review].count(1)
print('There are {} sentences end with .'.format(n))
```

### 將所有. 換成 ,

```python
all_review = [i.replace('.', ',', -1) for i in all_review]
print('\n'.join(all_review[:20]))
```

### 將所有sentence 中的第一個 the 置換成 The

```python
all_review = [i.replace('the', 'The') for i in all_review]
print('\n'.join(all_review[:20]))
```

### 將偶數句子全部轉換為大寫，基數句子全部轉換為小寫

```python
all_review = [i.upper() if all_review.index(i) % 2 == 0 else i.lower() for i in all_review]
print('\n'.join(all_review[:20]))
```

### 將所有句子合併在一起，並以' / ' 為間隔

```python
print('/'.join(all_review))
```
