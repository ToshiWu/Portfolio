## 運用scikit-learn API 實現K-fold分割資料

---

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
```

### 讀入資料

```python
dataset = pd.read_csv(r'./D19_Social_Network_Ads.csv')
display(dataset)
```

### 取出訓練特徵與標註

```python
X = dataset[['User ID', 'Gender', 'Age', 'EstimatedSalary']].values
Y = dataset['Purchased'].values
```

---

### 將訓練資料按照順序切割成10等分

```python
kf = KFold(10)
kf.get_n_splits(X)
print(kf)
```

### 將訓練資料隨機切割成10等分

```python
kf = KFold(10, shuffle=True)
kf.get_n_splits(X)
print(kf)
```

---

### 取出 切割資料對應位置

```python
train_split = kf.split(X)
next(train_split)
```

### Or

```python
for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "\nTEST:", test_index,
          sep = '\n',end = '\n'*4)
```

### 取出切割資料：trainset / testset 特徵(x_train/x_test)/標註(y_train/y_test)
